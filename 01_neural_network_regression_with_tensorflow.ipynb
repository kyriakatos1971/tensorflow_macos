# Introduction to Regression with Neural Networks in Tensorflow

```python
# Import Tensorflow
import tensorflow as tf
print(tf.__version__)
```

```python
# Creating some data for view and fit
import numpy as np
import matplotlib.pyplot as plt

# Create features

X = np.array([-7.0,-4.0,-1.0,2.0,5.0,8.0,11.0,13.0])

# Create labels 

y = np.array([3.0,6.0,9.0,12.0,15.0,18.0,21.0,23.0])


# Visualize it
plt.scatter(X,y);
```

## Steps in modelling with Tensorflow

1. **Creating a model** - define the input and the output layers, as well as the hidden layers of a deep learning moddel.
2. **Compiling a model** - define the loss functions and the optimizer and evaluation metrics
3. **Fitting a model**

```python
# Set random seed
tf.random.set_seed(42)


# 1. Create a model using the Sequential API
model = tf.keras.Sequential([tf.keras.layers.Dense(1)])

# 2. Compitle the model
model.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.SGD(), metrics = ["mae"])

# 3. Fit the model

model.fit(tf.expand_dims(X, axis=-1),y, epochs=1400)

```

```python
# Check out X and y

X, y
```

```python
# Try and make a prediction

model.predict([17.0])
```

# Improving our model

1. **Creating a model** -  add more layers, incr. number of hiddedn units, change the activation function of each layer
2. **Compiling a model** - change the optimization function or its learning rate
3. **Fitting a model** - fit a model for more **epochs** or on more data


```python
# Let's rebuild our model

# 1. Create a model (specified to your problem)
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(1)])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.SGD(), metrics = ["mae"])

# 3. Fit the model

model.fit(tf.expand_dims(X, axis=-1),y, epochs=100)


model.predict([17.0])
```

```python
# Let's rebuild our model - Take 2

# 1. Create a model (specified to your problem)
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(1)])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.SGD(), metrics = ["mae"])

# 3. Fit the model

model.fit(tf.expand_dims(X, axis=-1),y, epochs=500)


# Try and make a prediction

model.predict([17.0])
```

```python
# Let's rebuild our model - Take 3

# 1. Create a model (specified to your problem)
model = tf.keras.Sequential([
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(100, activation="relu"),
  tf.keras.layers.Dense(1)])

# 2. Compile the model
model.compile(loss=tf.keras.losses.mae, optimizer = tf.keras.optimizers.Adam(lr=0.001), metrics = ["mae"])

# 3. Fit the model

model.fit(tf.expand_dims(X, axis=-1),y, epochs=500)


# Try and make a prediction

model.predict([17.0])
```
